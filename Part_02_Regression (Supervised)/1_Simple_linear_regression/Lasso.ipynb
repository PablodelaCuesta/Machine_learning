```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns

# specified libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# seaborn style
sns.set()

# seed
np.random.seed(42)
```

```python
# datos sintéticos
X = 5 + 1.8 * np.random.randn(200)
res = 0.9 * np.random.randn(200)
y_obs = 5 + 1.9 * X + res

# visualizamos
plt.plot(X ,y_obs, marker='.', linestyle='none')
plt.xlabel("Valores X")
plt.ylabel("Valores y")
plt.show()
```

```python
# Reshape the data
X_s = np.reshape(a=X, newshape=(-1, 1))
y_obss = np.reshape(a=y_obs, newshape=(-1, 1))
```

```python
from sklearn.linear_model import Lasso
```

```python
reg = Lasso(alpha=0.1)
reg.fit(X_s, y_obss)
```

```python
y_pred = reg.predict(X_s)
```

```python
# Visualising the Training set results
plt.figure(figsize=(9,5))
plt.plot(X_s, y_obss, marker='.', linestyle='none', label='Observado')
plt.plot(X_s, y_pred, label='Teórico')
plt.legend()
plt.show()
```
